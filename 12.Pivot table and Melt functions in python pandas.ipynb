Reshaping of Python file

Melt function

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],"english":[12,1,3,21,4,56],"maths":[6,5,4,3,2,1]})
var
                
```

```python
pd.melt(var)
```

```python
#to make any variable as id
pd.melt(var,id_vars="days")
```

```python
pd.melt(var,id_vars="days",var_name="PYHTON")
```

```python
pd.melt(var,id_vars=["days"],var_name="PYHTON",value_name="PANDAS")
```

PIVOT FUNCTION

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],"st_names":["fatima","hadia","alina","fatima","hadia","alina"],"english":[12,1,3,21,4,56],"maths":[6,5,4,3,2,1]})
var
          
```

```python
var.pivot(index=["days"],columns="st_names")
```

```python
var.pivot(index=["days"],columns="st_names",values="english")
```

```python
var=pd.DataFrame({"days":[1,1,1,1,5,6],"st_names":["fatima","hadia","alina","fatima","hadia","alina"],"english":[12,1,3,21,4,56],"maths":[6,5,4,3,2,1]})
var
       
```

```python
var.pivot_table(index=["st_names"],columns="days",aggfunc="mean")
```
