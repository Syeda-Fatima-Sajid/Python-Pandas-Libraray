```python
import pandas as pd
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var1,var2,on="A")
```

```python
#merge onlythose  common in A
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,9],"C":[21,22,23,24]})

pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

#pd.merge(var1,var2,how="right")
pd.merge(var1,var2,how="left")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="outer")
```

```python
#to check which data is present and which is not
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var1,var2,how="outer",indicator=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var1,var2,how="outer",left_index=True,right_index=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var1,var2,how="outer",left_index=True,right_index=True,suffixes=("name","ID"))
```

```python
sr1=pd.Series([1,2,3])
sr2=pd.Series([4,5,6])
pd.concat([sr1,sr2])
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([var1,var2])
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B1":[21,22,23,24]})
pd.concat([var1,var2])
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B1":[21,22,23,24]})
pd.concat([var1,var2],axis=1)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2],"B1":[21,22]})
pd.concat([var1,var2],axis=1)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2],"B1":[21,22]})
pd.concat([var1,var2],axis=1,join="outer")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2],"B1":[21,22]})
pd.concat([var1,var2],axis=1,join="inner")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"B1":[21,22,23,24]})
pd.concat([var1,var2],axis=1,keys=['var1','var2'])
```
